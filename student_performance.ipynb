```python
df["avg_previous_grade"] = (df["G1"] + df["G2"]) / 2
df["Weekalc"] = (df["Dalc"] + df["Walc"]) / 2
features = ["studytime","Weekalc","schoolsup","absences","avg_previous_grade"]
target = "G3"

X = df[features]
Y = df[target]
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
X = X.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Before scaling: split first!!!
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

scaler = StandardScaler()

# Fit only on training data
X_train_scaled = scaler.fit_transform(X_train)

# Transform test data with the same scaler
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(Y_test, Y_pred)
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
r2 = r2_score(Y_test, Y_pred)

print(mae, rmse, r2)
```

```python
# 1. Train/test split BEFORE scaling
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

# 2. Scale numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 3. BASELINE MODEL (linear regression)
baseline_model = LinearRegression()
baseline_model.fit(X_train_scaled, y_train)
baseline_preds = baseline_model.predict(X_test_scaled)

baseline_mae = mean_absolute_error(y_test, baseline_preds)
baseline_rmse = np.sqrt(mean_squared_error(y_test, baseline_preds))
baseline_r2 = r2_score(y_test, baseline_preds)

print("Baseline Model:")
print("MAE:", baseline_mae)
print("RMSE:", baseline_rmse)
print("R²:", baseline_r2)

# 4. IMPROVEMENT: Polynomial features
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)

# IMPORTANT: fit only on training data
X_train_poly = poly.fit_transform(X_train_scaled)
X_test_poly = poly.transform(X_test_scaled)

# 5. Train polynomial regression model
poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)

# 6. Predictions with polynomial model
poly_preds = poly_model.predict(X_test_poly)

poly_mae = mean_absolute_error(y_test, poly_preds)
poly_rmse = np.sqrt(mean_squared_error(y_test, poly_preds))
poly_r2 = r2_score(y_test, poly_preds)

print("\nPolynomial Model (degree=2):")
print("MAE:", poly_mae)
print("RMSE:", poly_rmse)
print("R²:", poly_r2)
```
